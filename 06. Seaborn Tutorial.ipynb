```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
# plotting with only one parameter. Example, x='flipper_length_mm' as in our example.

# We can use the distribution modules functions to plot our data by using: histplot(), kdeplot(), ecdfplot() and rugplot()
# figure-level functions: displot(), jointplot() and pairplot()

x = sns.displot(
    
    # As a recommendation, we should let data=penguins first, so that we don't need to write a long code when we
    # we wanna call out each parameters of each data.
    # Example of long code writing:
        # a = penguins['flipper_length_mm'] (not letting data=penguins)
        # a = 'flipper_length_mm' (if we let data=penguins)
    data=penguins, y='flipper_length_mm',
    
    # kind='kde': change kind of shape from bar to curve.
    # kind='ecdf': change kind of shape from bar to ecdf (shape like ladder).
    # kind='kde': histogram shape (bars).
    kind='hist', 
    
    # kde=True or False: to show curve string on our graph.
    kde=False, 
    
    # rug=True or False: to show individual observations with a marginal.
    rug=True,
    
    # hue='species': show a small box that contain species and divide our histogram into each parts.
    # Each kind of plot can be drawn separately for subsets of data using hue mapping.
    hue='species', 
    
    multiple='stack',
    
    # From 'col' on, we called 'FacetGrid':
    # col='sex': provide 2 different graphs, one for male and another for female.
    col='sex', 
    
    # height and aspect: use to make the graphs have a better look with a better shape.
    height=4, aspect=1
);

# set_axis_labels() & set_titles(): Add titles and labels to each graph
x.set_axis_labels("Density (a.u)", "Flipper length (mm)");
x.set_titles("{col_name} penguins");

# *** remember that: col and hue are flexible according to our data.
```

```python
# plotting with only 2 parameter.

x = sns.displot(
    data=penguins, 
    x='flipper_length_mm', y='bill_length_mm',
    kind='kde',
    rug=True,
    hue='species',
    col='sex',
    height=4,
    aspect=1
);

x.set_axis_labels("Density (a.u.)", "Flipper length (mm)");
x.set_titles("{col_name} penguins");
```

```python
# plotting with only 2 parameter.

x = sns.displot(
    data=penguins, 
    x='flipper_length_mm', y='bill_length_mm',
    kind='hist',
    rug=True,
    hue='species',
    col='sex',
    height=4,
    aspect=1
);

x.set_axis_labels("Density (a.u.)", "Flipper length (mm)");
x.set_titles("{col_name} penguins");
```

### sns.jointplot( )
    - seaborn.jointplot(
          *, 
          x=None, 
          y=None, 
          data=None, 
          kind='scatter', 
          color=None, 
          height=6, 
          ratio=5, 
          space=0.2, 
          dropna=False, 
          xlim=None, 
          ylim=None, 
          marginal_ticks=False, 
          joint_kws=None, 
          marginal_kws=None, 
          hue=None, 
          palette=None, 
          hue_order=None, 
          hue_norm=None, 
          **kwargs
      )
      
      - This function provides a convenient interface to the JointGrid class, with several canned plot kinds. This is intended to be a fairly lightweight wrapper; if you need more flexibility, you should use JointGrid directly.

```python
sns.jointplot(
    data=penguins,
    x='flipper_length_mm', y='bill_length_mm',
    hue='species',
    
    # kind='reg, resid, hex' doesn't spp with hue='species' but it spp with kind='kde, scatter, hist'
    kind='kde'
);
```

```python
sns.jointplot(
    data=penguins,
    x='flipper_length_mm', y='bill_length_mm',
    
    # marker: all markers that are showing inside the plotting picture area.
    # markers can be '+', '.', '^', 
    marker='',
    
    # s stands for 'size'.
    s=100,
    
    # bins=25: divide histogram into 25 plots.
    # fill=True or False.
    # if 'False' => no fill in those 25 plots.
    # if 'True' => fill in those 25 plots.
    marginal_kws=dict(bins=25, fill=True)
);
```

```python
sns.jointplot(
    data=penguins,
    x='bill_length_mm', y='flipper_length_mm', 
    height=8,    # height or size is the same because 'size' has been renamed to 'height'
    ratio=2,
    marginal_ticks=True
);


# heightnumeric : Size of the figure (it will be square).
# rationumeric : Ratio of joint axes height to marginal axes height.
```

```python
import seaborn as sns
import numpy as np
from itertools import product
# white, dark, whitegrid, darkgrid, ticks
sns.set(style='darkgrid');

tips=sns.load_dataset('tips')
color=sns.color_palette()[2]
g=sns.jointplot(
    "total_bill", 'tip', 
    data=tips, 
    kind='reg', 
    xlim=(0, 60),
    ylim=(0, 12),
    color='k',
    height=7  # 'size' parameter has been renamed to 'height'
);

# Clear the axes containing
g.ax_joint.cla()

# Generate some colors and markers
colors=np.random.random((len(tips), 3))
markers=['x', 'o', 'v', '^', '<']*100

# Plot each individual point separately
for i, row in enumerate(tips.values):
    g.ax_joint.plot(
        row[0], 
        row[1], 
        color=colors[i],
        marker=markers[i]
    )

g.set_axis_labels('total bill', 'tip', fontsize=16);
```

```python
#  import libraries
import matplotlib.pyplot as plt
from matplotlib import colors
import colorsys

# pick colors
colr = ['g', 'tab:blue', 'firebrick', 'xkcd:yellowish orange']

# plot color and display HEX, RGB, RGBA, and HLS codes
for i in colr:
    #  get color HEX, RGB and RGBA codes
    c_hex  = colors.to_hex(i)   #  HEX code
    c_rgb  = colors.to_rgb(i)   #  RGB code
    c_rgba = colors.to_rgba(i)  #  RGBA code
    
    r,g,b  = colors.to_rgb(i)            # red, green, blue
    c_hsv  = colorsys.rgb_to_hsv(r,g,b)  # hue, saturation, value

    #  plot hlines with color = i
    plt.figure(figsize = (10,1))
    plt.text(0,0.05, i, fontsize = 18)
    plt.hlines(0,0,10, color = i ,linestyles = 'solid', linewidth = 30)
    plt.axis('off')
    plt.show()

    # format tuples for output
    c_rgb  = tuple([float("{0:.6f}".format(n)) for n in c_rgb])
    c_rgba = tuple([float("{0:.6f}".format(n)) for n in c_rgba])
    c_hsv  = tuple([float("{0:.6f}".format(n)) for n in c_hsv])

    # print COLOR, HEX, RGB and RGBA codes
    print('\tcolor:\t{}\n\thex:\t{}\n\trgb:\t{}\n\trgba:\t{}\n\thsv:\t{}'.format(i,c_hex, c_rgb, c_rgba, c_hsv))
```
