## class seaborn.JointGrid(**kwargs)

- Grid for drawing a bivariate plot with marginal univariate plots.
- Many plots can be drawn by using the figure-level interface jointplot(). Use this class directly when you need more flexibility.



```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

penguis = sns.load_dataset('penguins')
sns.JointGrid(data=penguis, x='bill_length_mm', y='bill_length_mm');
```

```python
g = sns.JointGrid(data=penguis, x='bill_length_mm', y='bill_depth_mm');
g.plot(sns.scatterplot, sns.histplot);

sns.set(style='darkgrid');
```

```python
g = sns.JointGrid(data=penguis, x='bill_length_mm', y='bill_depth_mm')
g.plot(
    sns.scatterplot, 
    sns.histplot, 
    alpha=.7,   #  alpha is in range from 0 to 1.
    edgecolor='1',
    linewidth=.5   # linewidth in the integer set.
);
```

```python
g = sns. JointGrid(data=penguis, x='bill_length_mm', y='bill_depth_mm')
g.plot_joint(sns.scatterplot, s=100, alpha=.7);
g.plot_marginals(sns.histplot, kde=True);
```

```python
# set up the grid without assigning any data
g = sns.JointGrid()
```

## JointGrid

- You can then plot by accessing the ax_joint, ax_marg_x, and ax_marg_y attributes, which are matplotlib.axes.Axes objects:
    

```python
g = sns.JointGrid()
x, y = penguis['bill_length_mm'], penguis['bill_depth_mm']

sns.scatterplot(
    x=x, y=y, ec='b', fc='none', s=100, 
    linewidth=1.5, ax=g.ax_joint
)

sns.histplot(x=x, fill=False, linewidth=2, ax=g.ax_marg_x)
sns.kdeplot(y=y, linewidth=2, ax=g.ax_marg_y)
```

```python
g = sns.JointGrid(data=penguis, x='bill_length_mm', y='bill_depth_mm')
g.plot(sns.regplot, sns.boxplot);
```

```python
g = sns.JointGrid(data=penguis, x='bill_length_mm', y='bill_depth_mm', 
                  hue='species', height=10, ratio=5)
g.plot(sns.scatterplot, sns.histplot, alpha=1);
```
